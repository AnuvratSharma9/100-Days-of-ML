## Ordinal Encoder

### Ordinal encoding is a way to convert categorical variables (categories/labels) into numbers, but it preserves the order or ranking of the categories.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\DELL\Downloads\customer.csv")
df.head(5)
```

```python
df=df.iloc[:,2:]
df
```

# Splitting the Dataset  

```python
from sklearn.model_selection import train_test_split
X=df.drop('purchased',axis=1)
y=df['purchased']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)

```

# Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
oe.fit(X_train)
X_train = oe.transform(X_train)
X_test= oe.transform(X_test)
X_train
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le.fit(y_train)
y_train=le.transform(y_train)
y_test=le.transform(y_test)
y_test

# Label Encoding is for targeted inputs
```
